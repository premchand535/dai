```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "/kaggle/input/top-100-badminton-players/flashscore.csv"  
df = pd.read_csv(file_path)

# Select a column for univariate analysis
column = "Points"  
# Convert to numeric if needed
df[column] = pd.to_numeric(df[column], errors='coerce')

# Drop NaN values (if any) after conversion
df = df.dropna(subset=[column])

# Summary statistics
print("Summary Statistics:")
print(df[column].describe())

# Histogram
plt.figure(figsize=(6, 4))
sns.histplot(df[column], bins=20, kde=True)
plt.title(f"Histogram of {column}")
plt.show()

# Boxplot (Fix applied)
plt.figure(figsize=(6, 4))
sns.boxplot(x=df[column])
plt.title(f"Boxplot of {column}")
plt.show()

# KDE Plot
plt.figure(figsize=(6, 4))
sns.kdeplot(df[column], shade=True)
plt.title(f"KDE Plot of {column}")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "/kaggle/input/top-100-badminton-players/flashscore.csv"  # Change this to your dataset path
df = pd.read_csv(file_path)

# Selecting numerical columns
numerical_columns = df.select_dtypes(include=["number"]).columns

# Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Pair plot
sns.pairplot(df[numerical_columns])
plt.show()

# Scatter plot between two variables (change column names accordingly)
plt.figure(figsize=(6, 4))
plt.show()
```
